```python
##4 逆序
# start:end:-1 表示从start开始逆序读取至end结束（不包含end），索引号 end <= start  
```

```python
# 考虑一个 0~9 的简单序列向量， 逆序取到第 1 号元素，不包含第 1 号
# 创建 0~9 向量
x = tf.range(9) 
# 从 8 取到 0，逆序，不包含 0
x[8:0:-1]
```

```python
# 逆序全部元素
x[::-1] 
```

```python
# 逆序间隔采样
x[::-2] 
```

```python
##5 使用...避免冒号过多的情况（示例在下方）
```

```python
# 创建4维张量（4张彩色图片）
x = tf.random.normal([4,32,32,3])
```

```python
# 读取第 2,3 张图片
x[1:3]
```

```python
# 读取第一张图片
x[0,::] 
```

```python
x[:,0:28:2,0:28:2,:]
```

```python
# 读取每张图片的所有通道，其中行按着逆序隔行采样，列按着逆序隔行采样
```

```python
x = tf.random.normal([4,32,32,3])
# 行、列逆序间隔采样
x[0,::-2,::-2] 
```

```python
# 取 G 通道数据 
x[:,:,:,1] # 当张量的维度数量较多时，不需要采样的维度一般用单冒号:表示采样所有元素
```

```python
# 读取最后 2 张图片
# 高、宽、通道维度全部采集，等价于 x[2:]
x[2:,...] 
```

```python
# 读取所有样本，所有高、宽的前 2 个通道（R和G）的数据
x[...,:2] 
```

```python
# 读取第 1~2 张图片（所有高、宽的）后 2 个通道（G和B）的数据
x[0:2,...,1:] 
```
